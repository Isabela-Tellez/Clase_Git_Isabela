### **La distribución Normal: La tiranía del Promedio**

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Gráfico campana de Gauss
alturas = np.random.normal(loc = 170, scale = 10, size = 1000)          #loc es media / Altura máx, scale desviación estándar / datos que se separaban los datos del centro (centro entendido como estado optimo), size datos aleatorios que se van a generar en la muestra
plt.figure(figsize = (10, 3))
sns.histplot(alturas, kde = True, color = "orange")  #Kde muestra la línea
#Son datos continuos

#Titulo del gráfico
plt.title("Distribución Normal: Alturas")

#Indicadores del gráfico
plt.legend("leyenda") #la fina es el kde, y la gruesa son las barras del gráfico

#Añadir texto o etiquetas en los ejes (x, y)
plt.xlabel("Alturas (cm)")
plt.ylabel("Frecuencia")

plt.show()
```

## **La Distribución Binomial: "Cara o Cruz"**

Es el éxito o el fracaso repetido muchas veces.


- **n**: cuántos intentos  
- **p**: probabilidad de éxito en cada intento  
- Resultado: **cuántos éxitos** obtuviste

Si lanzamos una moneda 10 veces, ¿cuál es el número más probable de 'caras' que obtendréis? (¡Exacto, el 5!)".

```python
# n = intentos, p = probabilidad de éxito, k = número de éxitos
monedas = np.random.binomial(n=10, p=0.5, size=1000)
plt.figure(figsize=(8,4))
sns.histplot(monedas, color="salmon", discrete=True)
plt.title("Distribución Binomial: Lanzamiento de Monedas")
plt.xlabel("número de caras")
plt.ylabel("Frecuencia")
plt.show()
```

## **La Distribución de Poisson: "La Espera"**

Nos sirve para contar cuántas veces ocurre algo en un tiempo determinado:


**λ (lambda)**: el promedio de eventos por intervalo  
- Resultado: un conteo: 0, 1, 2, 3…


Ejemplo real: llegan correos a una bandeja de entrada.  
Promedio: 2 correos por 1 hora → λ=2.

```python
#Gráfica
# lam = promedio esperado de eventos
#generamos los datos
emails = np.random.poisson(lam=2, size=1000)
plt.figure(figsize=(8,4))#tamaño de figura
#discrete=True le dice a Seaborn: "estos datos solo toman valores enteros, centra una barra en cada uno". Es la versión cómoda del truco -0.5.
sns.histplot(emails, color="purple", discrete=True)
plt.title("Distribución de Poisson: Emails por hora")
plt.xlabel("Emails recibidos por hora")
plt.ylabel("Frecuencia")
plt.show()
```

## **Ejercicio guiado práctico de Distribución Binomial (éxitos en intentos)**

**Ejemplo real:** 30 personas ven una landing.
Probabilidad de compra p=0.2.

¿Cuántas compras esperas?

```python
#Visualización en Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
n = 30
p = 0.2

# Simulación
compras = np.random.binomial(n=n, p=p, size=1000)

# Gráfico con Matplotlib
plt.figure(figsize=(8,4))
plt.hist(compras, bins=range(min(compras), max(compras)+2), color='skyblue', edgecolor='black')

plt.title("Distribución Binomial (Matplotlib)")
plt.xlabel("Número de compras")
plt.ylabel("Frecuencia")

plt.show()
```

```python
#Visualización en Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
sns.histplot(compras, color="salmon", discrete=True)

plt.title("Distribución Binomial (Seaborn)")
plt.xlabel("Número de compras")
plt.ylabel("Frecuencia")

plt.show()
```

### Ejercicio 5 – Ajuste rápido de parámetros (sin dolor)
- Normal: μ ≈ mean, σ ≈ std  
- Poisson: λ ≈ mean  
- Binomial: p ≈ mean/n (con n=50)  
Interpreta qué significa cada parámetro.

```python
#EJERCICIO 4
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos
tiempos_entrega = np.random.normal(loc=25, scale=3, size=2000)
tiempos_mixtos = np.concatenate([
    np.random.normal(loc=25, scale=3, size=1800),
    np.random.normal(loc=45, scale=5, size=200)
])

# Histograma comparativo
plt.figure(figsize=(10,5))
sns.histplot(tiempos_entrega, color='skyblue', bins=40, label='Normal', kde=True)
sns.histplot(tiempos_mixtos, color='salmon', bins=40, label='Mixta', kde=True, alpha=0.6)
plt.title("Comparación de tiempos de entrega: Normal vs Mixta")
plt.xlabel("Tiempo de entrega")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

```python
##EJERCICIO 5

#Normal
mu = np.mean(tiempos_entrega)
sigma = np.std(tiempos_entrega)
print(f"Normal: μ ≈ {mu:.2f}, σ ≈ {sigma:.2f}")

#Poisson
llegadas_ds = np.random.poisson(lam=5, size=2000)
lambda_poisson = np.mean(llegadas_ds)
print(f"Poisson: λ ≈ {lambda_poisson:.2f}")

#Binomial
conversiones_ds = np.random.binomial(n=50, p=0.12, size=2000)
p_binomial = np.mean(conversiones_ds)/50
print(f"Binomial: p ≈ {p_binomial:.3f}")
```

### Ejercicio 7 – Probabilidad aproximada (Poisson)
Con λ=5:
1) Probabilidad de 0 llegadas (simulación)  
2) Probabilidad de ≥10 (simulación)  
3) Interpreta para planificación (3–5 líneas)

```python
#EJERCICIO 6
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros
n_visitas = 100
p_A = 0.10
p_B = 0.13
simulaciones = 1000

# Simulación
conversiones_A = np.random.binomial(n=n_visitas, p=p_A, size=simulaciones)
conversiones_B = np.random.binomial(n=n_visitas, p=p_B, size=simulaciones)

# Visualización
plt.figure(figsize=(10,5))
sns.histplot(conversiones_A, color='skyblue', discrete=True, label='Versión A', alpha=0.6)
sns.histplot(conversiones_B, color='salmon', discrete=True, label='Versión B', alpha=0.6)
plt.title("Simulación A/B: Conversiones en 100 visitas")
plt.xlabel("Número de conversiones")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# ¿Cuántas veces B gana a A?
B_gana_A = np.sum(conversiones_B > conversiones_A)
print(f"B gana a A en {B_gana_A} de {simulaciones} experimentos")
```

```python
#EJERCICIO 7
# Parámetros
lambda_poisson = 5
simulaciones = 100000

# Simulación
llegadas = np.random.poisson(lam=lambda_poisson, size=simulaciones)

# Probabilidad de 0 llegadas
prob_0 = np.mean(llegadas == 0)

# Probabilidad de >= 10 llegadas
prob_10_mas = np.mean(llegadas >= 10)

print(f"Probabilidad de 0 llegadas: {prob_0:.4f}")
print(f"Probabilidad de 10 o más llegadas: {prob_10_mas:.4f}")
```
